```
plot = VectorPlot[{y, Sin[x]}, {x, 2, 2}, {y, 5, 5},
  VectorColorFunction -> None,
  VectorScaling -> Automatic,
  VectorStyle -> Directive[Gray, Opacity[0.3]]]
```

```
plot1 = StreamPlot[{y, Sin[x]}, {x, 2, 2}, {y, 5, 5},
  StreamScale -> None,
  StreamColorFunction -> None,
  StreamPoints -> {{{{2, 0}, Green}, {{-3, -1}, Green}, Automatic}},
  StreamStyle -> Directive[Blue, Opacity[0.5]]]
```

```
isoclines = ContourPlot[-Sin[x]==0*y, {x, 2, 2}, {y, 5, 5},
  ContourStyle -> Directive[Red, Opacity[0.4]]]
```

```
example = Labeled[
  Show[plot, plot1, isoclines],
  LineLegend[{Directive[Gray, Opacity[0.3]], Green, Directive[Blue, Opacity[0.5]], Directive[Red, Opacity[0.4]]}, {"Campo vectorial", "Trayectorias", "Diagrama de fase", "Isoclinas"}],
  {Right}
]
```

```
Export["Example.pdf", example]
```
