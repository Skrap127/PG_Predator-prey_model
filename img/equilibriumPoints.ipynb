```
plot = VectorPlot[{-x, 2y}, {x, 3, 3}, {y, 3, 3},
  Frame -> False,
  PlotTheme -> "Scientific",
  VectorColorFunction -> None,
  VectorStyle -> Directive[Gray, Opacity[0.6]]]
```
