```
plot1 = StreamPlot[{y, -x + (1-x^2)y}, {x, 5, 5}, {y, 5, 5},
  StreamColorFunction -> None,
  StreamPoints -> {{{{2.0002, 0.2}, Red}, Automatic}}
]
```

```
plot2 = StreamPlot[{y, -x + -(1-x^2)y}, {x, 5, 5}, {y, 5, 5},
  StreamColorFunction -> None,
  StreamPoints -> {{{{2.0002, 0.2}, Red}, Automatic}}
]
```

```
ciclos = GraphicsRow[{Labeled[plot1, "Ciclo lmite estable", Top], Labeled[plot2, "Ciclo lmite inestable", Top]}]
```

```
Export["CicloLimEstInest.pdf", ciclos]
```
